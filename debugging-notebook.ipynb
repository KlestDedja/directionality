```python
import pandas as pd
import matplotlib.pyplot as plt
import os

root_folder = os.getcwd()

df = pd.read_csv(os.path.join(root_folder, "images-3D-lightsheet-20240928_BAM_fkt20_P3_fkt21_P3_PEMFS", "HOG_stats_None_64pixels_clean.csv"))

# Extract the two columns
col_mode_duration = "mode direction"  # fix spelling if needed
col_std_dev_mode = "std. deviation (mode)"

# Drop missing rows that could cause errors
df.dropna(subset=[col_mode_duration, col_std_dev_mode], inplace=True)

# Plot a scatter plot for visual correlation
plt.scatter(df[col_mode_duration], df[col_std_dev_mode], alpha=0.7, s=10)
plt.axvline(x=90, linestyle='--')  # Add vertical line at x=90
plt.xlabel(col_mode_duration)
plt.ylabel(col_std_dev_mode)
plt.title("Correlation Scatter Plot")
plt.show()
```
